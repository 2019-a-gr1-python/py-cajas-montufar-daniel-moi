### 5) Obtener los titulos de las columnas especificas

```python
def readinputdata(CARS_PATH):
       fichero=open(CARS_PATH,'r')
       f=[]
       line='0'
       while len(line)>0:
        line=array(fichero.readline().split()).astype(float)
        if len(line)>0 :
            f.append(line)
       fichero.close()
       return array(f)
```

### 18) Distribucion de capacidad de carga

```python
# Primero se obtienen el top de autos nuevos
autos_mas_nuevos = cars['Año de la puesta (Generación)'].sort_values(ascending=False).head(1000)
anio_autos_mas_nuevos = cars.loc[autos_mas_nuevos.index,['id_trim','Marca','Año de la puesta (Generación)', 'Velocidad máxima [km/h]']]
print('TOP 1000:')
display(anio_autos_mas_nuevos)
```

### 19) Histograma de vehiculos por anio 

```python
data = anio_autos_mas_nuevos['Año de la puesta (Generación)'].values
print(data.shape)
(n, bin_edges) = np.histogram(data, bins=50)
plt.plot(bin_edges[1:],n)
plt.show()
```

### 21) Histograma de vehicuos por potencia del motor
